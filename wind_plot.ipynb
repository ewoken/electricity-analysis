```python
from ipyleaflet import Map, TileLayer, basemaps, FullScreenControl
from ipyleaflet.velocity import Velocity
import xarray as xr
```

```python
ds = xr.load_dataset('./_data/copernicus/europe_cold_wave_1985.grib', engine="cfgrib")
```

```python
da = ds.sel(time='1985-01-03').mean(dim='time')
da
```

```python
center = [45, 8]
zoom = 5
m = Map(center=center, zoom=zoom, interpolation='nearest', basemap=basemaps.CartoDB.DarkMatter)
display_options = {
    'velocityType': 'Global Wind',
    'displayPosition': 'bottomleft',
    'displayEmptyString': 'No wind data'
}
wind = Velocity(data=da,
                zonal_speed='u100',
                meridional_speed='v100',
                latitude_dimension='latitude',
                longitude_dimension='longitude',
                #velocity_scale=0.01,
                max_velocity=15,
                display_options=display_options)
m.add_layer(wind)

m.add_control(FullScreenControl())
m
```
